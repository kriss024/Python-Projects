## IBM HR Analytics Employee Attrition & Performance
---
### Projekt zaliczeniowy na przedmiot Big Data

Analiza czynników wpływających na rotację pracowników w firmie IBM.<br/>Zbiór danych pochodzi ze strony [kaggle.com](https://www.kaggle.com/pavansubhasht/ibm-hr-analytics-attrition-dataset/data#)

```
hr_dane = spark.read.format('com.databricks.spark.csv').options(inferSchema="true", header='true', delimiter =',').load('/FileStore/tables/HR_Analytics_Employee_Attrition.csv')
display(hr_dane)
```

```
print('Liczba wierszy:', hr_dane.count(), '\nLiczba kolumn:', len(hr_dane.columns))
```

Przed dalszą analizą warto sprawdzić typy znajdujących się w tabeli zmiennych.

```
hr_dane.printSchema()
```

Warto nadać zmiennym odpowiedni typ, np. kategorie, aby zwiększyć czytelność analizy

```
mapping= {
'1': 'Below College',
'2': 'College',
'3': 'Bachelor',
'4': 'Master',
'5': 'Doctor'
    }
hr_dane2 = hr_dane.withColumn('Education', hr_dane['Education'].cast("string"))
hr_dane2 = hr_dane2.replace(to_replace=mapping, subset=['Education'])

mapping= {
'1': 'Low',
'2': 'Medium',
'3': 'High',
'4': 'Very High'
}
hr_dane2 = hr_dane2.withColumn('EnvironmentSatisfaction', hr_dane2['EnvironmentSatisfaction'].cast("string"))
hr_dane2 = hr_dane2.replace(to_replace=mapping, subset=['EnvironmentSatisfaction'])

mapping= {
'1': 'Low',
'2': 'Medium',
'3': 'High',
'4': 'Very High'
}
hr_dane2 = hr_dane2.withColumn('JobInvolvement', hr_dane2['JobInvolvement'].cast("string"))
hr_dane2 = hr_dane2.replace(to_replace=mapping, subset=['JobInvolvement'])

mapping= {
'1': 'Low',
'2': 'Medium',
'3': 'High',
'4': 'Very High'
}  
hr_dane2 = hr_dane2.withColumn('JobSatisfaction', hr_dane2['JobSatisfaction'].cast("string"))
hr_dane2 = hr_dane2.replace(to_replace=mapping, subset=['JobSatisfaction'])

mapping= {
'1': 'Low',
'2': 'Good',
'3': 'Excellent',
'4': 'Outstanding'
}
hr_dane2 = hr_dane2.withColumn('PerformanceRating', hr_dane2['PerformanceRating'].cast("string"))
hr_dane2 = hr_dane2.replace(to_replace=mapping, subset=['PerformanceRating'])

mapping= {
'1': 'Low',
'2': 'Medium',
'3': 'High',
'4': 'Very High'
}
hr_dane2 = hr_dane2.withColumn('RelationshipSatisfaction', hr_dane2['RelationshipSatisfaction'].cast("string"))
hr_dane2 = hr_dane2.replace(to_replace=mapping, subset=['RelationshipSatisfaction'])

mapping= {
'1': 'Bad',
'2': 'Good',
'3': 'Better',
'4': 'Best'
}
hr_dane2 = hr_dane2.withColumn('WorkLifeBalance', hr_dane2['WorkLifeBalance'].cast("string"))
hr_dane2 = hr_dane2.replace(to_replace=mapping, subset=['WorkLifeBalance'])

display(hr_dane2)
```

```
hr_dane3 = hr_dane2.drop('EmployeeID')
hr_dane3 = hr_dane3.drop('EmployeeCount')
hr_dane3 = hr_dane3.drop('StandardHours')
hr_dane3 = hr_dane3.drop('Over18')
```

```
display(hr_dane3)
```

```
def job_level(x):
  if x == 1:
    return 'Junior Specialist'
  elif x==2:
    return 'Specialist'
  elif x==3:
      return 'Senior Specialist'
  elif x==4:
      return 'Expert'
  else:
    return 'Department Director'  
```

```
from pyspark.sql.types import *
udf_job_level = udf(job_level, StringType())
hr_dane4 = hr_dane3.withColumn('JobLevelDesc', udf_job_level(hr_dane3.JobLevel))
```

Struktura pracowników w departamentach

```
from pyspark.sql.functions import desc
hr_dane4.groupBy('JobLevelDesc').count().sort(desc('count')).show(truncate = False)
```

```
import matplotlib.pyplot as plt 
pandas = hr_dane4.groupBy('JobLevelDesc').count().sort(desc('count')).toPandas()

fig,ax = plt.subplots(1,1,figsize=(10,6))
pandas.plot(x = 'JobLevelDesc', y = 'count', kind = 'barh', color = 'C0', ax = ax, legend = False)
ax.set_xlabel('Job Level', size = 16)
ax.set_ylabel('# count', size = 16)
plt.show()
```

Badanie pracowników IBM było przeprowadzone w następujących departamentach

```
hr_dane4.select('Department').distinct().sort('Department').show(truncate = False)
```

W celu przyśpieszenia analizy wykorzystamy technologię Spark SQL oraz bazę danych HIVE

```
hr_dane4.createOrReplaceTempView("hr_dane_tmp")
spark.sql("DROP TABLE IF EXISTS hr_dane4")
spark.sql("CREATE TABLE hr_dane4 AS SELECT * FROM hr_dane_tmp")
```

### Analiza i wizualizacja danych

```
%sql

select Attrition, count(*)
from hr_dane4
group by Attrition;
```

```
%sql

select Education, EducationField, count(*)
from hr_dane4
group by Education, EducationField;
```

```
display(hr_dane4.groupBy('Attrition','Department').count())
```

```
display(hr_dane4.groupBy('JobSatisfaction','Department').count())
```

```
display(hr_dane4.groupBy('JobSatisfaction', 'JobRole').count().sort(desc("count")))
```

```
display(hr_dane4.groupBy('Attrition','OverTime').count())
```

Jak możemy zaobserwować na poniższym wykresie. Pracownicy świadczący pracę w nadgodzinach cechują się wyższym wskaźnikiem rotacji.

```
display(hr_dane4.groupBy('Attrition','OverTime').count())
```

Jak widzimy na poniższym zestawieniu pracownicy odchodzący z często posiadali niższe wynagrodzenie, często poniżej 3.000-4.000. Różnica w średnich dla obu grup to ok. 2.000.

```
%sql

select Attrition, mean(MonthlyIncome) as MonthlyIncome
from hr_dane4
group by Attrition;
```

```
display(hr_dane4.select('Attrition', 'MonthlyIncome').collect())
```

```
display(hr_dane4.groupBy('Attrition','TotalWorkingYears').count().sort('TotalWorkingYears'))
```

Poniższy wykres obrazuje, że najczęściej odchodzą z pracownicy ze stażem do 2 lat, lub z bardzo dużym stażem, jednak w drugim przypadku, możemy się spodziewać, że związane jest to z przejściem na emeryturę.

```
display(hr_dane4.groupBy('Attrition','TotalWorkingYears').count().sort('TotalWorkingYears'))
```

Jak możemy zobaczyć na wykresie dotyczącym stanowisk. Są pewne stanowiska gdzie wskaźnik rotacji jest wyższy.

```
display(hr_dane4.groupBy('Attrition', 'JobRole').count().sort(desc("count")))
```

Kolejnym czynnikiem wypływjącym na rotację jest wiek pracownika. Poniżej widać, że pracownicy poniżej 30 lat częściej zmieniają pracę.

```
display(hr_dane4.select('Attrition', 'Age').collect())
```

Na podstawie analizy oraz wizualizacji danych zostały zidentyfikowane następujące czynniki wpływające na rotację pracowników, są to praca w nadgodzinach (OverTime), staż pracy (TotalWorkingYears), zarobki (MonthlyIncome), zajmowane stanowisko (JobRole) oraz wiek pracownika (Age).
