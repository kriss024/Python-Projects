```python
columns=['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus']
prefix = ['BusTrav', 'Dept', 'Edu', 'Job', 'MarStat']

data_dummies = pd.get_dummies(data, columns = columns, prefix = prefix)
data_dummies.head()
```

```python
dict = {'Yes' : 1, 'No' : 0} 
data_dummies['Attrition'].replace(dict, inplace=True)
dict = {'Male' : 1, 'Female' : 0} 
data_dummies['Gender'].replace(dict, inplace=True)
data_dummies.head()
```

```python
data_dummies.describe()
```

```python
f = pd.DataFrame(data_dummies.dtypes)
f.columns = ['Data Types']
display(f)
```

```python
data_dummies_nonan = data_dummies.dropna()
```

```python
from sklearn import preprocessing
y_dummies = data_dummies_nonan.Attrition
y = pd.DataFrame(y_dummies)
x_dummies = data_dummies_nonan.drop(['Attrition'], axis=1)
l = x_dummies.columns
scaler = preprocessing.StandardScaler()
scaled_df = scaler.fit_transform(x_dummies)
X = pd.DataFrame(scaled_df)
X.columns = l
```

```python
X.reset_index(drop=True, inplace=True)
y.reset_index(drop=True, inplace=True)
data_std = pd.concat([y, X], axis=1)
data_std.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression(solver='lbfgs')
logisticRegr.fit(x_train, y_train.values.ravel())
```

```python
predictions = logisticRegr.predict(x_test)
proba = logisticRegr.predict_proba(x_test)
f = pd.DataFrame(proba)
f.describe()
```

```python
score = logisticRegr.score(x_test, y_test)
print(score)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
tn, fp, fn, tp = confusion_matrix(y_test, predictions).ravel()
tpr = tp/(tp+fn)
tnr = tn/(tn+fp) 
Accuracy = (tp+tn)/(tp+fp+fn+tn)
Youden = tpr + tnr - 1
y_scores = proba[:, 1]
AUC = roc_auc_score(y_test, y_scores)
Gini = AUC*2-1
sensitivity = tp/(tp+fn)
print("Accuracy: %s Youden: %s AUC: %s Gini: %s" % (Accuracy, Youden, AUC, Gini))
print("=> Sensitivity: %s" % sensitivity)
```

```python
y_scores = proba[:, 1]
thresh = 0.5
preds = np.where(y_scores>thresh, 1, 0)
tn, fp, fn, tp = confusion_matrix(y_test, preds).ravel()
sensitivity = tp/(tp+fn)
print("Sensitivity: %s for threshold: %s" % (sensitivity, thresh))
```

```python
Youden = 0
threshold = 0
for i in range(1, 100000):
    thresh = i / 100000
    preds = np.where(y_scores>thresh, 1, 0)
    tn, fp, fn, tp = confusion_matrix(y_test, preds).ravel()
    tpr = tp/(tp+fn)
    tnr = tn/(tn+fp)     
    J = tpr + tnr - 1
    if J > Youden:
        Youden = J
        threshold = thresh
        
print("Best Youden's J statistic: %s for threshold: %s" % (Youden, threshold))
```
