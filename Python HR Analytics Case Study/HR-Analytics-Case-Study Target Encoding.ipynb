```python
import pandas as pd  
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
%matplotlib inline
```

```python
dane = pd.read_csv("general_data.csv")
dane.head()
```

```python
dane.dropna(inplace = True)
```

```python
dane.describe(include=[np.object])
```

```python
dane.drop("Over18", axis=1, inplace=True)
```

```python
dane.describe(include=[np.number]) 
```

```python
dane.drop(["EmployeeID", "EmployeeCount", "StandardHours"], axis=1, inplace=True)
```

```python
c = dane["Attrition"].value_counts()
f = pd.DataFrame(c)
display(f)
```

```python
dict = {'Yes' : 1, 'No' : 0} 
dane['Attrition'].replace(dict, inplace=True)
```

```python
def Target_Encoding(x,y,data):
    agg = data.groupby(x)[y].mean()
    dict = agg.to_dict()
    return dict
```

```python
dict = Target_Encoding('BusinessTravel','Attrition',dane)
dane['BusinessTravel'].replace(dict, inplace=True)

dict = Target_Encoding('Department','Attrition',dane)
dane['Department'].replace(dict, inplace=True)

dict = Target_Encoding('EducationField','Attrition',dane)
dane['EducationField'].replace(dict, inplace=True)

dict = Target_Encoding('Gender','Attrition',dane)
dane['Gender'].replace(dict, inplace=True)

dict = Target_Encoding('JobRole','Attrition',dane)
dane['JobRole'].replace(dict, inplace=True)

dict = Target_Encoding('MaritalStatus','Attrition',dane)
dane['MaritalStatus'].replace(dict, inplace=True)

dane.head()
```

```python
dane.describe(include=[np.number])
```

```python
y = dane.Attrition
X = dane.drop(['Attrition'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression(solver='lbfgs')
logisticRegr.fit(x_train, y_train)
```

```python
predictions = logisticRegr.predict(x_test)
proba = logisticRegr.predict_proba(x_test)
f = pd.DataFrame(proba)
f.describe()
```

```python
score = logisticRegr.score(x_test, y_test)
print(score)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
cm = confusion_matrix(y_test, predictions)
print(cm)
```
